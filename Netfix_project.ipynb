```python
movie_review_count #movie id based on ratings
```

```python
benchmark=round(movie_review_count.quantile(0.6)) # taking movie threshold based on ratings whether to consider movie or not (or we can say it as outliers)
```

```python
drop_movie_index=movie_review_count[movie_review_count<benchmark].index
```

```python
drop_movie_index,drop_cust_index
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset['Cust_id'].isin(drop_cust_index)]
netflix_dataset=netflix_dataset[~netflix_dataset['movie_id'].isin(drop_movie_index)]
```
